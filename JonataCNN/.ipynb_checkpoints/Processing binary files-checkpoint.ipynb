## Preprocessing

From binary files to uniform and same sized .input files.

```python
# f = open('ACO.class.txt', 'r', encoding='utf-8', newline='')
# file_contents = f.read()
# print (file_contents)

max_character_encoding = 130 # it should be 126, but maybe there are characters that I didn't see in the training set
num_files = 32

def text_from_file(file):
    with open('./binaries/' + file, 'rb') as f:
        text = f.read()
        bin_text = text.decode('latin-1').encode("utf-8")
        str_text = str(bin_text)
        return str_text
    
def len_characters_file(file):
    return len(text_from_file(file))

def convert_str_to_array(text):
    print(text)
    content_numbers = np.zeros(len(text) )
    for i in range(len(text)):
#             print(ord(str_text[i]) / max_character_encoding)
            content_numbers[i] = ord(text[i]) / max_character_encoding
    return content_numbers
```

Getting the size of the largest binary files.

```python
import os
import numpy as np
import math

big_array = []
Y = np.zeros([1, num_files])
file_names = [""]*num_files
count = 0

for root, dirs, files in os.walk("./binaries"):
    for file in files:
        if file.endswith(".txt"):
            file_names[count] = file
            
            # creating the new file
            text_file = text_from_file(file)
            new_array = convert_str_to_array(text_file)
            big_array.append(new_array)
            #print("    size: ", len(new_array))
            
            if file.endswith(".o.txt"):
                Y[0,count] = 1
            count = count + 1
            
            # saing a temporary file
#             with open(file + ".temp", 'w') as f:
#                 for line in text_file:
#                     f.write(line)

Y = Y.transpose()
print(Y.shape)
```

Defining the size of the input.

```python
max_size_row = -1
num_files = 0

for r in big_array:
    num_files = num_files + 1         
    l = len(r)
    if(l > max_size_row):
        max_size_row = l

max_size_row = max_size_row + 1000 # adding just in case the testing set has more characters that are not in the training set
            
print("max_size_row: " + str(max_size_row))
print("num_files: " + str(num_files))
```

Making all the files of the same size.

```python
import os

# The size of a row (one per compiled file)
height = int( math.sqrt( int(  max_size_row / 3 ) ) )
width = height
total_size = height * width * 3


X = np.empty([num_files, total_size])

count = 0
for row in big_array:
    # print(len(row))
    X[count] = np.pad(row, (0, total_size - len(row)), 'constant')
    # print(X[count])
    count = count + 1

X.shape
```

Create the numpy X and Y with all the files pre-processed and their corresponding labels.

## Using Tensorflow to classify the files

```python
import math
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import imread
import scipy
from PIL import Image
import pandas as pd
import tensorflow as tf
import tensorflow.keras.layers as tfl
from tensorflow.python.framework import ops

%matplotlib inline
np.random.seed(1)
```

```python
print(X.shape)
print(Y.shape)
```

```python
def show_compiled_as_image(index):
    height = int( math.sqrt( int(  len(big_array[index]) / 3 ) ) )
    width = height
    total_pixels = height * width * 3

    #print("X[index] = ", len(X[index]))
    #print("total_pixels = ", total_pixels)

    test = X[index,:total_pixels]

    test = np.reshape(test, (height, width, 3))

    #print(test.shape)
    print(file_names[index].replace(".txt",""))
    plt.imshow(test) #display sample training image
    plt.show()

for i in range(10):
    show_compiled_as_image(i)
```

Shuffling and breaking in train and test set.

```python
randomize = np.arange(len(file_names))
np.random.shuffle(randomize)

X_shuffled = X[randomize]
Y_shuffled = Y[randomize]
names_randomized = np.array(file_names)[randomize]

size_train = 25

X_train = X_shuffled[:size_train]
X_train = np.reshape(X_train, (size_train, height, width, 3))

X_test  = X_shuffled[size_train:]
X_test = np.reshape(X_test, (len(file_names)-size_train, height, width, 3))

Y_train = Y_shuffled[:size_train]
Y_test  = Y_shuffled[size_train:]

#print(Y_shuffled)
#print(names_randomized)
```

Creating my model.

```python
def my_model():
    """
    Implements the forward propagation for the binary classification model:
    ZEROPAD2D -> CONV2D -> BATCHNORM -> RELU -> MAXPOOL -> FLATTEN -> DENSE
    
    Arguments:
    None

    Returns:
    model -- TF Keras model (object containing the information for the entire training process) 
    """
    model = tf.keras.Sequential([
            
        ## ZeroPadding2D with padding 3, input shape of height x width x 3
        tfl.ZeroPadding2D(padding=3,input_shape=(height, width, 3)) ,
        
        ## Conv2D with 32 7x7 filters and stride of 1
        tfl.Conv2D(filters=32, kernel_size=7, strides=1) ,
        
        ## BatchNormalization for axis 3
        tfl.BatchNormalization(axis=3) ,
        
        ## ReLU
        tfl.ReLU(),
        
        ## Max Pooling 2D with default parameters
        tfl.MaxPooling2D(),
        
        ## Flatten layer
        tfl.Flatten(),
        
        ## Dense layer with 1 unit for output & 'sigmoid' activation
        tfl.Dense(units=1, activation='sigmoid')

        ])
    
    return model
```

Defining the optimizar, loss function and metric.

```python
my_model = my_model()
my_model.compile(optimizer='adam',
                 loss='binary_crossentropy',
                 metrics=['accuracy'])
my_model.summary()
```

```python
my_model.fit(X_train, Y_train, epochs=30, batch_size=num_files)
```

```python
my_model.evaluate(X_test, Y_test)
```
